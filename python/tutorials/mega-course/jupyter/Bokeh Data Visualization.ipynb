```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
#import io
#import requests

output_file("WomenEngineers.html")

url="http://pythonhow.com/data/bachelors.csv"
#s=requests.get(url).content
#df = pd.read_csv(io.StringIO(s.decode('utf-8')))

df = pd.read_csv(url)

x = df["Year"]
y = df["Engineering"]

f = figure(plot_width=500, plot_height=400, tools='pan')
f.title.text="Women Engineers"
f.title.text_color="Red"
f.title.text_font='arial'
f.title.text_font_style='italic'
f.title.text_font_style='bold'
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Year"
f.yaxis.axis_label="Percentage"

f.line(x,y)
show(f)
```

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
import io, requests

output_file("PressureOverTemp.html")

url="http://pythonhow.com/verlegenhuken.xlsx"
df = pd.read_excel(url)
x = df["Temperature"]/10
y = df["Pressure"]/10

f = figure(plot_width=1000, plot_height=1000, tools='pan, reset')
f.title.text="Pressure over Temperature"
f.title.text_color="Green"
f.title.text_font='arial'
f.title.text_font_style='italic'
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Temperature"
f.yaxis.axis_label="Pressure (hPa)"

f.circle(x,y)
show(f)
```

```python
from bokeh.plotting import figure, output_file, show

output_file("Scatter_plotting.html")

p = figure(plot_width=500, plot_height=400, tools = 'pan, reset')
p.title.text = "Earthquakes"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.yaxis.minor_tick_line_color = "Yellow"
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"

p.circle([1,2,3,4,5], [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color="red", alpha=0.5)
p.line([i*2 for i in range(1,11)], range(10), line_width=2, color="blue")

show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

output_file("ADBE.html")

df = pandas.read_csv("adbe.csv", parse_dates=["Date"])

p = figure(width=1000, height=700, x_axis_type="datetime")
p.title.text = "Adobe"
p.title.text_font_style = "bold"
p.title.text_color = "Red"
p.yaxis.axis_label = "Closing Price"

p.line(df["Date"], df["Close"], color="Red", alpha=0.5)
show(p)
```
