```python
from bs4 import BeautifulSoup
import requests, pandas as pd

url = "http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s={}.html"
l = []

r = requests.get(url.format(str(0)), headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
soup = BeautifulSoup(r.content, "html.parser")
page_num = soup.find_all("a", {"class":"Page"})[-1].text

for page in range(0, int(page_num)*10, 10):
    url.format(str(page))

    r = requests.get(url, headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    properties = soup.find_all("div", {"class":"propertyRow"})

    for item in properties:
        d = {}
        price = item.find("h4", {"class":"propPrice"}).text.strip() #replace("\n","").replace(" ", "")
        #print(price)
        d["Price"] = price

        address = item.find_all("span", {"class":"propAddressCollapse"})
        #print(address[0].text + ", " + address[1].text)
        d["Address"] = address[0].text + ", " + address[1].text

        num_beds = item.find("span", {"class":"infoBed"})
        if num_beds:
            #print(num_beds.text)
            d["Beds"] = num_beds.text
        else:
            d["Beds"] = None

        sqft = item.find("span", {"class":"infoSqFt"})
        if sqft:
            #print(sqft.text)
            d["SqFt."] = sqft.text
        else:
            d["SqFt."] = None

        full_baths = item.find("span", {"class":"infoValueFullBath"})
        if full_baths:
            #print(full_baths.text)
            d["Full Baths"] = full_baths.text
        else:
            d["Full Baths"] = None

        half_baths = item.find("span", {"class":"infoValueHalfBath"})
        if half_baths:
            #print(half_baths.text)
            d["Half Baths"] = half_baths.text
        else:
            d["Half Baths"] = None

        for column_group in item.find_all("div", {"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span", {"class":"featureGroup"}), column_group.find_all("span", {"class":"featureName"})):
                if "Lot Size" in feature_group.text:
                    #print(feature_group.text, feature_name.text)
                    d["Lot Size"] = feature_name.text
        l.append(d)
    #print(" ")

df = pd.DataFrame(l)
#df.to_csv("rocksprings_wy.csv")
#df1 = pd.read_csv("rocksprings_wy.csv")
df
```
